```python
data_subset = df.label.value_counts(normalize=False, sort=True).to_dict()  # normalize returns the amount of samples
print(f'Label value counts \n{data_subset}')
plt.figure(0, figsize=(10,10))
plt.pie(data_subset.values(), labels=data_subset.keys(), autopct="%1.1f%%")
plt.title("Label Distribution In The Dataset")
L=plt.legend()
L.get_texts()[0].set_text('0 = no predator in conversation')
L.get_texts()[1].set_text('1 = predator in conversation')
L.get_texts()[2].set_text('2 = watermarked conversation')
plt.show()
```

```python
%%time
plt.figure(figsize=(9, 9))
# sns.heatmap(df.corr(), cmap='coolwarm',annot = True)
sns.heatmap(df.corr(), cmap='coolwarm',annot = False)
plt.title('Correlation matrix of all the features')
```

```python
def find_features_with_highest_corr_to_label(df:pd.DataFrame, threshold:float=0.5) -> pd.DataFrame:
    corrs = df.corr()
    labels_row = corrs['label']
    return labels_row[labels_row > np.abs(threshold)].index.to_list()
```

```python
threshold=0.45
features_with_highest_corr_to_label = find_features_with_highest_corr_to_label(df=df, threshold=threshold)
df_highest_correlation = df[features_with_highest_corr_to_label]
```

```python
%%time
plt.figure(figsize=(10, 10))
sns.heatmap(df_highest_correlation.corr(), cmap='coolwarm',annot = True)
plt.title(f'Correlation matrix of features with over threshold {threshold} to the label')
```

# PCA

```python
%%time
from sklearn.decomposition import PCA
n_components = 20

X_train = df.drop(columns=['label'])

print(f'Performing PCA on train data')
pca = PCA(n_components=n_components).fit(X_train)
```

```python
variance_ratio = pca.explained_variance_ratio_  #assumption: this is already sorted (importent for the cumsum)
cumulative_sum = np.cumsum(variance_ratio)
plt.figure(0, figsize=(16,8))
plt.plot(cumulative_sum)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.xticks(np.arange(0, len(cumulative_sum)+1, 1.0))  
plt.title('Cumulative explained variance after PCA')
plt.grid()
plt.show()
```
